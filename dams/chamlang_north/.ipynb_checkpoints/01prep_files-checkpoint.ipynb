```python
import os
from glob import glob
import geopandas as gpd
from pathlib import Path
from typing import List, Union
from osgeo import gdal, gdal_array
```

```python
data_path = './data'
```

```python
# convert units of range offsets from pixels to radians 
fn_list = glob('data/*/*/*/S1*rng_off.tif')
for fn in fn_list:
    out_fn = f'{fn[:-4]}_phonyunw.tif'
    !gdal_calc.py -A $fn --calc "(A*2.3)*(12.5663706/0.05546576)" --outfile=$out_fn --hideNoData --NoDataValue=0 --overwrite
```

```python
def clip_hyp3_products_to_common_overlap(data_path: Union[str, Path], overlap: List[float]) -> None:
    """Clip all GeoTIFF files to their common overlap
    
    Args:
        data_dir:
            directory containing the GeoTIFF files to clip
        overlap:
            a list of the upper-left x, upper-left y, lower-right-x, and lower-tight y
            corner coordinates of the common overlap
    Returns: None
    """
    
    files_for_mintpy = ['_phonyunw.tif', 'phonyconncomp.tif', '_conncomp.tif', '_corr.tif', '_unw_phase.tif', '_dem.tif', '_lv_theta.tif', '_lv_phi.tif']

    for extension in files_for_mintpy:
        print(f'working on {extension}') 
        for file in data_path.rglob(f'*{extension}'):

            dst_file = file.parent / f'{file.stem}_aoi{file.suffix}'

            gdal.Translate(destName=str(dst_file), srcDS=str(file), projWin=overlap)
```

```python
# get overlap from aoi
region_fn = '../../../../mapping/hunku_region.shp'
region_gdf = gpd.read_file(region_fn)
aoi_overlap = [region_gdf.bounds.minx.item(),
               region_gdf.bounds.maxy.item(),
               region_gdf.bounds.maxx.item(),
               region_gdf.bounds.miny.item()]
```

```python
# clip files to aoi
clip_hyp3_products_to_common_overlap(Path(data_path), aoi_overlap)
```
